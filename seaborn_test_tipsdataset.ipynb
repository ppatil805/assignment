```python
 import seaborn as sns
v=sns.load_dataset('tips')
v.head(5)
```

```python
sns.histplot(data=v['tip'])
```

```python
sns.histplot(data=v['tip'],bins=[1,2,3,4])
```

Insights:From the above hist plot we know the how the tips  count varies of the it shows high at perticular days only.

```python
sns.jointplot(data=v,x='total_bill',y='tip')
```

Insights: form the above graph we analys the relation between total bill and tips per day and it shows the linear relation of both.as the tips increases the bill value also increases.

```python
sns.jointplot(data=v,x='total_bill',y='tip',kind='hex')
```

Insights:this plot also shows the linear relation of total bill and tips per day and at leftside near bottom the custers get dark shows accumulation of data and not much changes in bill value and tip value.

```python
sns.jointplot(data=v,x='total_bill',y='tip',kind='reg')
```

Insights: Regration  graph shows the relation between total bill and tips per day and it shows the linear relation of both.as the tips increases the bill value also increases.

```python
sns.jointplot(data=v,x='total_bill',y='tip',hue='sex')
```

Insights: form the above graph we analys the relation between total bill and tips per day,as the tips increases the bill value also increases but if we can see considering sex i.e. who give more tips so from this graph it is clear that male gives more tipsa than female

Insights:From the above KDE plot shows the probability density oftotal bill,the pick shows that maximum bill in between 10 to 20.
