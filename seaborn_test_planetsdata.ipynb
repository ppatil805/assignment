```python
import seaborn as sns
pl=sns.load_dataset('planets')
pl.head(5)
```

```python
sns.histplot(data=pl['mass'])
```

```python
sns.histplot(data=pl['mass'],bins=[0,1,2,3,4,5])
```

Insights:this hist plot shows the relationship of mass of planets with respect to counts

```python
sns.jointplot(data=pl,x='mass',y='orbital_period')
```

Insights: This jointplot is mass vs orbital period but from the plot we can't get any usefull or specific information .

```python
sns.jointplot(data=pl,x='mass',y='orbital_period',kind='hex')
```

Insights: This jointplot is mass vs orbital period the area of highdensity shows the relation in between but from the plot we can't get any usefull or specific information .

```python
sns.jointplot(data=pl,x='mass',y='orbital_period',kind='reg')
```

```python
sns.jointplot(data=pl,x='mass',y='orbital_period',hue='year')
```

```python
sns.kdeplot(data=pl['orbital_period'])
```

Insights: from the above plot we can conclude that most of the planets have short orbital period around 100 days and some planets have longer orbit period.

```python
sns.pairplot(data=pl[['mass','orbital_period']])
```

```python
sns.pairplot(pl,hue='year')
```

Insights: from the above plot the year of origin of planet maks not clear impact on different variales .
