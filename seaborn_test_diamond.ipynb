```python
import seaborn as sns
d=sns.load_dataset('diamonds')
d.head(5)
```

```python
sns.histplot(data=d['carat'])
```

```python
sns.histplot(data=d['carat'], bins = [0.3,1,2,3])
```

insights:From the above histogram  we can conclude that the carat count from 0.3 to 1 is mostly in demand and then carat count from 1 to 2
has some what less in demand and the carat count from 2 to 3.

```python

sns.jointplot(data=d,x='carat',y='price')

```

Insights:This plot is shows the linear relation between price and carat
as the carat is increases the price also increases.

```python
sns.jointplot(data=d,x='carat',y='price',kind='hex')
```

Insights:This plot is shows the linear relation between price and carat as the carat is increases the price also increases

```python
sns.jointplot(data=d,x='carat',y='price',kind='reg')
```

Insights:Regration plot shows the linear relation between price and carat as the carat is increases the price also increases

```python
sns.jointplot(data=d,x='carat',y='price',hue='cut')
```

This plot is shows the visual relashion between price and carat with different cuts in diamonds, from this plot we can see the how the price and carat plot varies with different cut and shown with diferent color.

```python
sns.kdeplot(data=d)
```

```python
sns.kdeplot(data=d['carat'])
```

Insights: this plot shows the how price of diamond varies.

```python
sns.pairplot(data=d[['carat','price']])
```

```python
sns.pairplot(data=d[['carat','price','cut']],hue='cut')
```

insights:this plot shows price vs carat relationship with all aspects like cuts in the diamond etc overall all the subplots shows the linear relation of the parameters .
